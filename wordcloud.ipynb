# fetching topics cleaning and generating word cloud

```python
subreddit = reddit.subreddit('telangana')

# Fetch hot trending topics
hot_topics = subreddit.hot(limit=100)  # You can adjust the limit as needed
text_data = ""

for submission in hot_topics:
    text_data += submission.title + " "

# Generate the word cloud
wordcloud = WordCloud(width=800, height=800,
                      background_color='white',
                      min_font_size=10).generate(text_data)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
